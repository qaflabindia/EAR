# Credit Risk Guru — a Ksetra runtime for personal-loan underwriting

This notebook builds one concrete, runnable EAR (Enterprise Agentic Runtime)
instance — **`Credit-Risk-Guru-Ksetra`** — and walks an applicant through it
end to end.

Credit risk underwriting is a good showcase for EAR because the domain
already insists on the same separations EAR is built around:

| EAR concern | Credit risk equivalent |
|---|---|
| `Dharma` policy gates | Underwriting policy thresholds (min credit score, DTI ceiling, fair-lending limits) |
| `Pramana` evidence | The "why" a regulator/auditor demands behind every adverse-action or approval decision |
| `Smriti` memory | The applicant/portfolio decision log |
| `Anubhava` experience | Portfolio-level approval-rate patterns |
| `Samskara` adaptation | A distilled underwriting lesson fed back into future reasoning |
| `Pariksha` checker layer | A "maker-checker" / dual-control gate on every pipeline stage's output |

No LLM credentials are required — everything below runs on EAR's
dependency-free default reasoning path, the same path the README documents
for running with no Manas attached. A commented-out section shows where a
real model would plug in, without ever putting a key in this notebook.

## 0. Imports

Pull in the philosophical-stack classes (`Sankalpa` → `Vidya` → `Guna` →
`Varna` → `Karma` → `Dharma` → `Ksetra`) plus `Pariksha`, which we'll poke at
directly later on.

````python
import os

from ear import Dharma, Guna, Karma, Ksetra, Pariksha, Sankalpa, Varna, Vidya

print("EAR philosophical stack ready.")
````

## 1. `Vidya` — the skills a credit analyst actually uses

A `Vidya` is one addressable capability. We give our persona three small,
inspectable scoring skills instead of one opaque "score the applicant"
black box — each is independently testable and swappable.

````python
def credit_score_tier(score: int) -> str:
    if score >= 720:
        return "prime"
    if score >= 660:
        return "near-prime"
    if score >= 620:
        return "subprime"
    return "deep-subprime"


def dti_tier(debt_to_income_ratio: float) -> str:
    if debt_to_income_ratio <= 0.30:
        return "low"
    if debt_to_income_ratio <= 0.45:
        return "moderate"
    return "high"


_RISK_GRID = {
    ("prime", "low"): "A", ("prime", "moderate"): "A", ("prime", "high"): "B",
    ("near-prime", "low"): "B", ("near-prime", "moderate"): "B", ("near-prime", "high"): "C",
    ("subprime", "low"): "C", ("subprime", "moderate"): "C", ("subprime", "high"): "D",
    ("deep-subprime", "low"): "D", ("deep-subprime", "moderate"): "D", ("deep-subprime", "high"): "E",
}


def risk_grade(score_tier: str, dti_band: str) -> str:
    return _RISK_GRID.get((score_tier, dti_band), "E")


score_skill = Vidya(name="credit_score_tier", description="Bands a FICO score", handler=credit_score_tier)
dti_skill = Vidya(name="dti_tier", description="Bands a debt-to-income ratio", handler=dti_tier)
grade_skill = Vidya(name="risk_grade", description="Combines score + DTI tiers into a risk grade", handler=risk_grade)

print([s.name for s in (score_skill, dti_skill, grade_skill)])
````

## 2. `Guna` — the "Credit Risk Guru" persona

A `Guna` stacks `Vidya` skills behind a behavioural nature and standing
instructions — the persona that actually carries out the underwriting
judgement.

````python
credit_risk_guru = Guna(
    name="Credit Risk Guru",
    instructions=(
        "Underwrite personal loans conservatively. Band every applicant's "
        "credit score and debt-to-income ratio, then assign a risk grade "
        "before any approval reasoning happens."
    ),
)
credit_risk_guru.add_skill(score_skill).add_skill(dti_skill).add_skill(grade_skill)

print(credit_risk_guru.name, "->", [s.name for s in credit_risk_guru.skills])
````

## 3. `Varna` — the underwriting workflow

A `Varna` orders one or more `Guna` personas into a workflow. Here it's a
single-persona workflow; a larger desk might stack a fraud-review persona
ahead of the credit-risk persona in the same `Varna`.

````python
underwriting_workflow = Varna(name="Underwriting Workflow")
underwriting_workflow.add_persona(credit_risk_guru)

print(underwriting_workflow.name, "->", [p.name for p in underwriting_workflow.personas])
````

## 4. `Karma` — the process

A `Karma` is the executable process: one or more `Varna` workflows stacked
into an action a `Ksetra` can discover and run.

````python
loan_underwriting = Karma(name="Personal Loan Underwriting")
loan_underwriting.add_workflow(underwriting_workflow)

print(loan_underwriting.name, "->", [w.name for w in loan_underwriting.workflows])
````

## 5. `Dharma` — underwriting policy gates

Each `Dharma` is a governance rule, safely evaluated (no `eval`/`exec` — see
`ear/_safe_eval.py`) against the `Sankalpa`'s context. `Niyamana` checks
every one of these *before* anything else in the cycle runs, and stops the
cycle outright if any is violated — these are the bank's non-negotiable
underwriting and fair-lending floors, not soft suggestions.

````python
policies = [
    Dharma(name="Minimum Credit Score", rule="credit_score >= 620"),
    Dharma(name="Debt-to-Income Ceiling", rule="debt_to_income_ratio <= 0.45"),
    Dharma(name="No Active Defaults", rule="existing_defaults == 0"),
    Dharma(name="Loan Amount Cap", rule="loan_amount <= 75000"),
]

for policy in policies:
    print(f"- {policy.name}: {policy.rule}")
````

## 6. `Ksetra` — assemble Credit Risk Guru's runtime battlefield

`Ksetra` is where the process, the policies, and all nineteen pipeline
stages come together. We register the process and the policies; everything
else (`Niyamana`, `Anveshana`, `Varana`, `Samyojana`, `Niyojana`, `Pariksha`,
`Samanvaya`, `Smriti`, `Anubhava`, `Samskara`, ...) is already wired in by
`Ksetra`'s own `__init__`.

````python
runtime = Ksetra(name="Credit-Risk-Guru-Ksetra")
runtime.add_process(loan_underwriting)
for policy in policies:
    runtime.add_policy(policy)

print(runtime.name)
print("Processes:", [p.name for p in runtime.processes])
print("Policies: ", [p.name for p in runtime.policies])
````

## 7. `Manas` — optional: give the Guru a real mind

you want real LLM-backed reasoning instead, attach a `Manas`:


````python
if os.environ.get("OPENAI_API_KEY"):
    from ear import Manas

    runtime.manas = Manas(provider="openai", model="gpt-4o-mini", api_key=os.environ["OPENAI_API_KEY"])
    print("Manas attached -- Bhuddi will reason against a real LLM.")
else:
    print("No OPENAI_API_KEY in the environment -- Bhuddi uses its dependency-free default reasoning.")
````

## 8. Turning an applicant into a `Sankalpa`

In a real desk, feature computation (score banding, DTI banding, risk
grading) happens before the runtime ever reasons about the application.
`assess_applicant` invokes the Guru's own `Vidya` skills to compute those
features, folds them into the `Sankalpa`'s context, then hands the whole
thing to `runtime.reason()` — so the policy gates and the explanation both
see the same derived risk grade a human underwriter would.

````python
def assess_applicant(name: str, credit_score: int, debt_to_income_ratio: float,
                      annual_income: float, loan_amount: float, existing_defaults: int = 0):
    score_tier = credit_risk_guru.get_skill("credit_score_tier").invoke(credit_score)
    dti_band = credit_risk_guru.get_skill("dti_tier").invoke(debt_to_income_ratio)
    grade = credit_risk_guru.get_skill("risk_grade").invoke(score_tier, dti_band)

    context = {
        "credit_score": credit_score,
        "debt_to_income_ratio": debt_to_income_ratio,
        "annual_income": annual_income,
        "loan_amount": loan_amount,
        "existing_defaults": existing_defaults,
        "score_tier": score_tier,
        "dti_band": dti_band,
        "risk_grade": grade,
    }
    sankalpa = Sankalpa(
        text=f"Underwrite a ${loan_amount:,.0f} personal loan for {name} (risk grade {grade})",
        context=context,
    )
    return sankalpa


def show_decision(name: str, entry) -> None:
    print(f"Applicant: {name}")
    print(f"  Decision : {entry.decision}")
    print(f"  Basis    : {entry.evidence.basis}")
    print(f"  Plan     : {entry.evidence.sources['plan']}")
    print(f"  Policies checked: {entry.evidence.sources['policies_checked']}")
    print(f"  Explanation: {entry.evidence.sources['explanation']}")
````

## 9. Applicant A — clean approval

A prime-tier, low-DTI applicant comfortably clears every `Dharma` gate.

````python
sankalpa_a = assess_applicant(
    name="Asha Rao", credit_score=760, debt_to_income_ratio=0.22,
    annual_income=95_000, loan_amount=20_000, existing_defaults=0,
)
runtime.reason(sankalpa_a)
show_decision("Asha Rao", runtime.smriti.working[-1])
````

## 10. Applicant B — stopped by `Niyamana` before anything else runs

This applicant's debt-to-income ratio breaches the 0.45 ceiling. `Niyamana`
checks every `Dharma` policy *first* — before `Anveshana` even discovers a
process — and `Ksetra.reason()` raises `PermissionError` immediately. No
decision is reasoned, no `Smriti` entry is written: a policy violation never
quietly becomes a yes.

````python
sankalpa_b = assess_applicant(
    name="Vikram Shah", credit_score=700, debt_to_income_ratio=0.52,
    annual_income=80_000, loan_amount=25_000, existing_defaults=0,
)
try:
    runtime.reason(sankalpa_b)
except PermissionError as exc:
    print("Rejected before reasoning even started:")
    print(" ", exc)
````

## 11. `Pariksha` — the checker layer for every maker stage

Past the `Dharma` gate, four "maker" stages each produce something that the
next stage has to trust: `Anveshana` discovers candidate processes, `Varana`
selects among them, `Samyojana` composes their workflows into a plan, and
`Niyojana` schedules that plan. `Pariksha` is the one place all four outputs
— plus `Nirnaya`'s final decision — get checked, instead of each maker stage
re-implementing its own validation.

To see it actually intervene rather than just sit there unused, we
temporarily make `Anveshana` misbehave — returning a `Varna` workflow where
it's supposed to return `Karma` processes — and watch `Pariksha` catch it
mid-pipeline.

````python
broken_workflow = Varna(name="not-a-karma-process")
original_discover = runtime.anveshana.discover
runtime.anveshana.discover = lambda *args, **kwargs: [broken_workflow]

try:
    runtime.reason(assess_applicant(
        name="Test Applicant", credit_score=700, debt_to_income_ratio=0.30,
        annual_income=70_000, loan_amount=15_000, existing_defaults=0,
    ))
except TypeError as exc:
    print("Pariksha rejected Anveshana's malformed output:")
    print(" ", exc)
finally:
    runtime.anveshana.discover = original_discover  # restore the real maker stage

print("\nDirectly, Pariksha exposes one validator per maker stage:")
print(" ", [m for m in dir(Pariksha) if m.startswith("validate")])
````

## 12. A small portfolio — building real memory and experience

One decision in isolation doesn't show `Smriti`/`Anubhava`/`Samskara`
compounding. Run a borderline applicant plus a small portfolio through the
same runtime so several cycles accumulate.

````python
sankalpa_c = assess_applicant(
    name="Priya Nair", credit_score=625, debt_to_income_ratio=0.40,
    annual_income=48_000, loan_amount=10_000, existing_defaults=0,
)
runtime.reason(sankalpa_c)
show_decision("Priya Nair", runtime.smriti.working[-1])

portfolio = [
    dict(name="Daniel Ibe", credit_score=680, debt_to_income_ratio=0.35, annual_income=72_000, loan_amount=30_000),
    dict(name="Mei Lin", credit_score=740, debt_to_income_ratio=0.18, annual_income=110_000, loan_amount=50_000),
    dict(name="Carlos Diaz", credit_score=622, debt_to_income_ratio=0.44, annual_income=48_000, loan_amount=10_000),
    dict(name="Sara Kim", credit_score=705, debt_to_income_ratio=0.28, annual_income=89_000, loan_amount=40_000),
]

for applicant in portfolio:
    runtime.reason(assess_applicant(existing_defaults=0, **applicant))
    show_decision(applicant["name"], runtime.smriti.working[-1])
    print()

print(f"Total successful cycles observed by Anubhava: {runtime.anubhava.observations}")
````

## 13. Inspecting `Smriti`, `Anubhava` and `Samskara`

Six applicants have now cleared the runtime. `Anukulana`'s default
`adapt_every=5` means the fifth successful cycle triggered `SamskaraBank`
to distill one new lesson from `Anubhava`'s aggregated experience — not on
every single call, so adaptation doesn't spam one fresh impression per
decision.

````python
print("=== Smriti (what happened) ===")
print(runtime.smriti.context_window())

print("\n=== Anubhava (the pattern across repeated cycles) ===")
print(runtime.anubhava.summary())

print("\n=== Samskara (the distilled adaptation) ===")
if runtime.samskara.impressions:
    for impression in runtime.samskara.impressions:
        print(f"- {impression.name}: {impression.insight}")
else:
    print("No Samskara distilled yet.")
````

## 14. The full `Pramana` audit trail behind one decision

Every entry's `evidence` is a `Pramana` — separate from the decision itself
— recording which reasoning path was used, which policies were checked,
the composed plan, what `Smarana` recalled from memory, `Vyakhya`'s
explanation, and `Parishodhana`'s audit flag. This is the artifact a credit
risk model-governance review (or an adverse-action notice) would actually
want to see.

````python
last_entry = runtime.smriti.working[-1]
import json
print(json.dumps(
    {**last_entry.evidence.sources, "decision": last_entry.decision, "basis": last_entry.evidence.basis},
    indent=2, default=str,
))
````

## Why this maps cleanly onto credit risk governance

- **`Dharma` policy gates** mirror hard underwriting/fair-lending floors
  (minimum score, DTI ceiling, exposure caps) — `Niyamana` enforces them
  *before* any reasoning runs, so a breach can never be reasoned around.
- **`Pariksha`'s maker-checker layer** is the same "four-eyes" / dual-control
  principle credit risk operations already require of human underwriters,
  just applied to every stage of the automated pipeline (`Anveshana`,
  `Varana`, `Samyojana`, `Niyojana`, `Nirnaya`).
- **`Pramana`** gives every decision an evidence trail independent of the
  decision text itself — exactly what SR 11-7-style model governance and
  adverse-action notices require: not just *what* was decided, but
  defensibly *why*.
- **`Smriti` → `Anubhava` → `Samskara`** keep "what happened", "the pattern
  across the portfolio", and "the adaptation drawn from it" as three
  distinct layers, instead of one model silently drifting on raw history —
  which is precisely the kind of unmonitored drift credit risk model
  governance exists to catch.

Swap the dependency-free `Bhuddi` fallback for a real `Manas` + compiled
DSPy program, plug in your bank's actual policy thresholds, and this same
`Ksetra` is a real (if minimal) underwriting runtime — fully inspectable,
one named class per concern, at every stage.
